## Предобработка данных

### Смена названий колонок

Выделим новые имена колонок в константы для удобства автокомплита кода
(в т.ч. при передаче в обычные массивы).

```python
AIRPORT_DISTANCE_COLUMN = 'airport_distance'
BALCONIES_COLUMN = 'balconies'
CEILING_HEIGHT_COLUMN = 'ceiling_height'
CITY_CENTER_DISTANCE_COLUMN = 'city_center_distance'
DAYS_EXPOSED_COLUMN = 'days_exposed'
FLOORS_IN_BUILDING_COLUMN = 'floors_in_building'
FLOOR_COLUMN = 'floor'
IMAGES_COLUMN = 'images'
IS_APARTMENT_COLUMN = 'is_apartment'
IS_OPEN_PLAN_COLUMN = 'is_open_plan'
IS_STUDIO_COLUMN = 'is_studio'
KITCHEN_AREA_COLUMN = 'kitchen_area'
LIVING_AREA_COLUMN = 'living_area'
LOCATION_COLUMN = 'location'
PARKS_AROUND_3K_COLUMN = 'parks_around_3k'
PARK_DISTANCE_COLUMN = 'park_distance'
PONDS_AROUND_3K_COLUMN = 'ponds_around_3k'
POND_DISTANCE_COLUMN = 'pond_distance'
PRICE_COLUMN = 'price'
PUBLICATION_DATE_COLUMN = 'publication_date'
ROOMS_COLUMN = 'rooms'
TOTAL_AREA_COLUMN = 'total_area'
```

```python
df = df.rename(columns={
    'total_images': IMAGES_COLUMN,
    'last_price': PRICE_COLUMN,
    'total_area': TOTAL_AREA_COLUMN,
    'first_day_exposition': PUBLICATION_DATE_COLUMN,
    'rooms': ROOMS_COLUMN,
    'ceiling_height': CEILING_HEIGHT_COLUMN,
    'floors_total': FLOORS_IN_BUILDING_COLUMN,
    'living_area': LIVING_AREA_COLUMN,
    'floor': FLOOR_COLUMN,
    'is_apartment': IS_APARTMENT_COLUMN,
    'studio': IS_STUDIO_COLUMN,
    'open_plan': IS_OPEN_PLAN_COLUMN,
    'kitchen_area': KITCHEN_AREA_COLUMN,
    'balcony': BALCONIES_COLUMN,
    'locality_name': LOCATION_COLUMN,
    'airports_nearest': AIRPORT_DISTANCE_COLUMN,
    'cityCenters_nearest': CITY_CENTER_DISTANCE_COLUMN,
    'parks_around3000': PARKS_AROUND_3K_COLUMN,
    'parks_nearest': PARK_DISTANCE_COLUMN,
    'ponds_around3000': PONDS_AROUND_3K_COLUMN,
    'ponds_nearest': POND_DISTANCE_COLUMN,
    'days_exposition': DAYS_EXPOSED_COLUMN,
})
```

### Заполнение пропущенных данных

Посмотрим на процент пропущенных значений в колонках:

```python
missing_count = df.isna().sum()
missing_count = missing_count[missing_count > 0]
missing_count /= len(df) / 100
(missing_count
 .sort_values(ascending=False)
 .plot(kind='bar',
       title='Процент пропущенных значений'));
```

#### Пропущенный флаг апартаментов


На графике выше видно, что почти 90% записей не имеют проставленный флаг
апартаментов. С большой вероятностью все они являются жилым фондом (сравнительно
новое изобретение).

Поэтому заменим пропущенные значения в этой колонке на `False`.

```python
df[IS_APARTMENT_COLUMN] = df[IS_APARTMENT_COLUMN].fillna(False)
```

#### Пропущенное количество балконов

Почти 50% записей имеют пропуски в количестве балконов. Скорее всего, пропуск
означает отсутствие балкона. Исправим пропуски на `False`.

```python
df[BALCONIES_COLUMN] = df[BALCONIES_COLUMN].fillna(0)
```

#### Пропущенные парки

Попробуем поискать значения в расстоянии и количестве парков, которые можно
восстановить друг из друга.

```python
df[(df[PARK_DISTANCE_COLUMN].isna())
   & (df[PARKS_AROUND_3K_COLUMN] > 0)]
```

```python
df[(df[PARKS_AROUND_3K_COLUMN].isna())
   & ~(df[PARK_DISTANCE_COLUMN].isna())]
```

Вывод: по этим двум колонкам данные не восстановить. Люди не заполняли их, потому
что не знают точно. Заменять их нулем не выглядит правильным.

#### Пропущенные пруды

Попробуем поискать значения в расстоянии и количестве прудов, которые можно
восстановить друг из друга.

```python
df[(df[POND_DISTANCE_COLUMN].isna())
   & (df[PONDS_AROUND_3K_COLUMN] > 0)]
```

Вывод: по этим двум колонкам данные не восстановить. Люди не заполняли их,
потому что не знают точно. Заменять их нулем не выглядит правильным.

Добавим день недели, месяц и год публикации объявления отдельными колонками

```python
PUBLICATION_DAY_COLUMN = 'publication_day'
PUBLICATION_MONTH_COLUMN = 'publication_month'
PUBLICATION_YEAR_COLUMN = 'publication_year'

df[PUBLICATION_DAY_COLUMN] = pd.DatetimeIndex(df[PUBLICATION_DATE_COLUMN]).weekday + 1
df[PUBLICATION_MONTH_COLUMN] = pd.DatetimeIndex(df[PUBLICATION_DATE_COLUMN]).month
df[PUBLICATION_YEAR_COLUMN] = pd.DatetimeIndex(df[PUBLICATION_DATE_COLUMN]).year
```

```python
def get_floor_category(floor, floors_total) -> str:
    if floor == 1:
        return 'первый'
    if (not pd.isna(floors_total)
            and floor == floors_total):
        return 'последний'
    return 'другой'
```

```python
spb_df = (
    df[(df[LOCATION_COLUMN] == 'Санкт-Петербург')
       & (~df[CITY_CENTER_DISTANCE_COLUMN].isna())]
    .copy()
    .reset_index()
)

CITY_CENTER_DISTANCE_KM_COLUMN = 'city_center_distance_km'
spb_df[CITY_CENTER_DISTANCE_KM_COLUMN] = (
    spb_df[CITY_CENTER_DISTANCE_COLUMN].apply(lambda d: round(d / 1000, 0))
)
```

```python
spb_df.plot(
    x=CITY_CENTER_DISTANCE_KM_COLUMN,
    y=MEDIAN_CITY_CENTER_M2_PRICE_COLUMN,
    kind='scatter', grid=True, alpha=0.05)
plt.title('Связь цены с удаленностью от центра Петербурга')
plt.xlabel('расстояние от центра, км')
plt.ylabel('средняя цена за 1м²');
```
