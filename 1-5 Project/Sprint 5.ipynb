# Определение факторов, влияющих на успешность игры

Из открытых источников за 2016 год доступны исторические данные о продажах игр,
оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation).
Данные за год могут быть неполными.

Нужно выявить определяющие успешность игры закономерности. Это позволит сделать
ставку на потенциально популярный продукт и спланировать рекламные кампании.

```python
def read_file(name: str):
    # try open local file first then — if failed — open on server
    try:
        return pd.read_csv(f'./datasets/{name}.csv', sep=',')
    except FileNotFoundError:
        try:
            return pd.read_csv(f'/datasets/{name}.csv', sep=',')
        except Exception as e:
            print('Не удалось считать файл с данными на сервере', e, sep='\n')
            return None

    except Exception as e:
        print('Не удалось считать файл с данными на диске', e, sep='\n')
    return None
```

### Корреляция продаж с отзывами

Посмотрим на корреляцию между продажами и отзывами критиков

```python
def get_sales_corr(column):
    return (
        df[(df['year_released'] >= min_year)]
        .groupby('platform')
        [['sales_total', column]]
        .corr()
        .dropna()
        .sort_values('sales_total', ascending=False)
        .unstack()
        [column]
        ['sales_total']
        .sort_values(ascending=False)
        .rename(column)
    )
```

```python
def plot_corr(column):
    platforms = df_sales.columns
    platforms_df = df[
        (df['platform'].isin(platforms))
        & (df['year_released'] >= min_year)]

    fig, axs = plt.subplots(3, 3)
    fig.set_size_inches(12, 12)

    platforms_arr = np.reshape(platforms, (-1, 3))

    for i in range(platforms_arr.shape[0]):
        for j in range(platforms_arr.shape[1]):
            ax = axs[i, j]
            platform = platforms_arr[i, j]
            ax.set_title(platform)
            data = (
                platforms_df[platforms_df['platform'] == platform]
                [['sales_total', column]]
            )
            data.plot(kind='scatter', ax=ax, y='sales_total', x=column)
            ax.set(xlabel='оценка', ylabel='продажи, млн')
```

```python
plot_corr('score_critic')
```

## Гипотеза 1. Разница пользовательских рейтингов XOne vs PC

**Нулевая гипотеза:** средние рейтинги между этими двумя платформами равны.

**Альтернативная гипотеза:** средние рейтинги между XOne и PC различаются.

Мы знаем, что продажи сильно отличаются, поэтому не будем считать дисперсии
рейтингов одинаковыми.

```python
alpha = 0.01

df_relevant = df[(df['year_released'] >= min_year)]

results = st.ttest_ind(
    df_relevant[df_relevant['platform'] == 'XOne']['score_user'].dropna(),
    df_relevant[df_relevant['platform'] == 'PC']['score_user'].dropna(),
    equal_var=False)
```

```python
print(f'p-value: {results.pvalue:.2e}')

if results.pvalue >= alpha:
    print("Средние рейтинги между XOne и PC равны")
else:
    print("Нельзя опровергнуть гипотезу одинаковости рейтингов")
```

… и это логично — обе принадлежат майкрософту (иногда еще и с crossbuy между),
оптимизация часто одинаково (делая версию под xbox чаще всего делают и под ПК).

```python
df_relevant = df_relevant[df_relevant.platform.isin(['XOne', 'PC'])]
g = sns.catplot(y='platform', x='score_user', kind="box", data=df_relevant, orient='h', palette='pastel')
g.fig.set_figwidth(10)
g.fig.set_figheight(2)
```

III. Про регионы:
1. По платформам: в европе и америке лидирует текущее и, за ним, прошлое поколение консолей.
В америке примерно одинаково любят и сони (25%), и майкрософт (21%) — последних,
возможно, как местную компанию. В европе сони лидирует почти в три раза
по сравнению с МС (36% vs 13%). Япония — другой мир. Половину рынка заняла нинтендо.
2—4 место (40%) занимают консоли сони. Майкрософта в Японии, можно сказать, нет
от слова «совсем».
2. Видим, что в америке и европе любят одни и те же жанры (даже почти в одинаковых
пропорциях): 30% action, 25% shooter, 15% sport.
Япония продолжает поддерживать имидж другой вселенной, где больше всего (36%)
любят РПГ. Action на 2-м месте, но с долей такой же как и в америке-европе (29%).
3. Видим, что в америке и европе по возрастным ограничениям повторяется та же история, что и с жанрами: тот же порядок и почти в одинаковых пропорциях: 17+ (37%), любой (18—21%), 10+ лет (11-12%). В японии в лидерах 60% без рейтинга, потом 13+ (15%), любой возраст (11%) и 17+ (10). Думаю, если делать предположение, то если бы рейтинги были, в топе и в японии был бы 17+.
